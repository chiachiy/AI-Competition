```python
import csv
import copy
with open('foundamental_2014-2018.csv',newline='') as csvfile:
        rows=csv.reader(csvfile)
        foundmatenal_data=[]
        for row in rows:
            foundmatenal_data.append(row)
#         print(foundmatenal_data)
with open('strategy_results_train.csv',newline='') as csvfile:
    rows=csv.reader(csvfile)
    strategy_data=[]
    for row in rows:
        strategy_data.append(row)
#     print(strategy_data)
result_data= copy.deepcopy(strategy_data)

for i in range(1,len(foundmatenal_data)):
    for j in range(1,len(result_data)):
        if(result_data[j][0][0:11]==foundmatenal_data[i][0][0:11]):
            if(foundmatenal_data[i][0][17])=="4":    
                if((int(result_data[j][0][16:18])<=3) and (int(result_data[j][0][12:16])==(int(foundmatenal_data[i][0][12:16]))+1)):
                    for k in range(1,5):
                        result_data[j].append(foundmatenal_data[i][k])       
            elif (foundmatenal_data[i][0][17])=="1":
                if((int(result_data[j][0][16:18])<=6 and int(result_data[j][0][16:18])>=4) and (int(result_data[j][0][12:16])==int(foundmatenal_data[i][0][12:16]))):
                    for k in range(1,5):
                        result_data[j].append(foundmatenal_data[i][k])
            elif (foundmatenal_data[i][0][17])=="2":  
                if((int(result_data[j][0][16:18])<=9 and int(result_data[j][0][16:18])>=7) and (int(result_data[j][0][12:16])==int(foundmatenal_data[i][0][12:16]))):
                    for k in range(1,5):
                        result_data[j].append(foundmatenal_data[i][k])
            elif (foundmatenal_data[i][0][17])=="3":
                 if((int(result_data[j][0][16:18])<=12 and int(result_data[j][0][16:18])>=10) and (int(result_data[j][0][12:16])==int(foundmatenal_data[i][0][12:16]))):
                    for k in range(1,5):
                        result_data[j].append(foundmatenal_data[i][k])

print(result_data)

        
# print(int(result_data[1][0][16:18]))
# print(result_data[1][0][12:16])
# print(result_data[1][0][16:18])
# print(foundmatenal_data[1][0][12:16])
# print(foundmatenal_data[1][0][17])
```
