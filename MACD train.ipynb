```python
for w in range(0, len(company_index)):
    df = pd.read_excel(company_index[w]+".xlsx")
    inputs = {
    'open': df["open"],
    'high': df["high"],
    'low':  df["low"],
    'close': df["close"],
    'volume': df["volume"]
    }
    macd = abstract.MACD(df)
    willr = abstract.WILLR(df)
    rsi = abstract.RSI(df)
    macd0 = pd.DataFrame(macd["macd"], columns=["macd"])
    macdsignal = pd.DataFrame(macd["macdsignal"], columns=["macdsignal"])
    willr0 = pd.DataFrame(willr, columns=["willr"])
    rsi0 = pd.DataFrame(rsi, columns=["rsi"])
    pd.concat([df, macd0, macdsignal, willr0, rsi0], axis=1).to_excel(company_index[w]+".xlsx", \
                                              sheet_name = company_index[w], index=False)
    if(w%100==0):
        print(w)
```

```python
train_set = pd.read_excel("strategy_results_train.xlsx")
```

```python
import copy
```

```python
dataset = copy.deepcopy(train_set)
```

```python
temp = []
for i in range(0, dataset.shape[0]):
# for i in range(0, 10):
    if(i%1000==0):
        print(i, datetime.datetime.now())
    trade_company = dataset.iloc[i][0].split("_")[0]
    trade_date    = dataset.iloc[i][0].split("_")[1]
    df = pd.read_excel(trade_company+".xlsx")
    trade_date_datetime = datetime.date(int(trade_date[:4]), int(trade_date[4:]), 1) - datetime.timedelta(days=1)
    lower_index = 0
    for index in range(0, df.shape[0]):
        if(df["date"].iloc[index]< trade_date_datetime):
            lower_index = index
        else:
            break
#     print(list(df.iloc[lower_index, :]))
    temp.append(list(df.iloc[lower_index, :]))
```
